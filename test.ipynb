```python
from yandexgpt import YandexGPT
from weather_db import Weather
```

```python
weather_url = "https://archive-api.open-meteo.com/v1/archive"
params = {
	"latitude": 55.797858,
	"longitude": 37.693822,
	"start_date": "1980-01-01",
	"end_date": "2025-01-20",
	"hourly": ["temperature_2m", "rain", "snowfall", "wind_speed_10m", "wind_direction_10m"]
}
weather = Weather(url=weather_url, params=params)
data = weather.get_data()
data
```

```python
engine = weather.SQLite(data)
```

```python
api_key = "api-key"
folder_id = "folder-id"

gpt = YandexGPT(api_key=api_key, folder_id=folder_id, engine=engine)
gpt
```

```python
%load_ext autoreload
%autoreload 2

Q = ['Привет, что ты умеешь делать?', 'Какая погода была 1 ноября 2024 года?', 
     'Шел ли дождь 2 марта 2024?', 'А на следующий день?',
     'Какая была максимальная температура в августе 2023?', 
     'Какая погода была неделю назад?', 'Сколько снега всего выпало в январе 2022?', 'Какой день был самый жаркий в 2021 году?']

sqls = []

for q in Q:
    a = gpt.invoke(q)
    sqls.append(gpt.last_sql)
    print(f"Q: {q}")
    print(f"A: {a}")
    print(".......")
```

```python
for s in sqls:
    print(s)
    print("......")
```
